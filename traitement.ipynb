```python
#rassemblement par departements, régions et nationale
df1_dep = df1[df1['Niveau']=='Départements']
df1_reg = df1[df1['Niveau']=='Régions']
df1_fr = df1[df1['Niveau']== 'France']
```

```python
df1_dep.tail(100)
```

```python
df1_dep.isna().sum()
```

```python
# on retire les totaux et les ensembles
df1_dep_filt = df1_dep[~df1_dep.PATHOLOGIE.str.contains('TOTAL') 
                       & df1_dep['Tranche d\'age'].str.contains('Tous âges confondus') 
                       & ~df1_dep['SEXE'].str.contains('Ensemble')]
```

```python
df1_dep_filt.isna().sum()
```

```python
df1_dep_filt_toutage = df1_dep_filt[['ZONE', 'ANNEE', 'PATHOLOGIE', 'SEXE', 'Tranche d\'age','ind_freq']]
```

```python
df1_dep_filt_toutage.isna().sum()
```

```python
df1_dep_filt_toutage['ind_freq'].dtype
```

```python
df1_dep_filt_toutage.to_csv('tou_age_test.csv')
```

```python
df1_dep_filt_toutage.isna().sum()
```

```python
df1_dep_filt_toutage['ind_freq'] = pd.to_numeric(
    df1_dep_filt_toutage['ind_freq'],
    errors='coerce'   # met NaN si ce n'est pas convertible
)
```

```python
duplicate = df1_dep_filt_toutage.duplicated()
df1_dep_filt_toutage[duplicate]
```

```python
df1_dep_filt_toutage.isna().sum()
```

```python
df1_dep_filt_toutage.to_csv(
    "mon_fichier_debug.csv",
    index=False,
    na_rep="NaN"   # <--- force l’écriture "NaN"
)
```

```python
# !test! regroupement par maladie pour connaitre le nombre total de pathologies
df1_patho = df1_dep_filt_toutage[['PATHOLOGIE']]
test = df1_patho.groupby(['PATHOLOGIE'])
test.size()
# Nous avons moins de lignes pour la covid-19 car les premiers cas sont arrivés en 2020
# 179 pathologies à partir de 2020 et 178 avant
```

```python
pivot = df1_dep_filt_toutage.pivot_table(
    index="ZONE",
    columns=["ANNEE", "SEXE", "PATHOLOGIE", "Tranche d\'age"],
    values="ind_freq",
    aggfunc="sum"
)
```

```python
pivot.transpose()
```
